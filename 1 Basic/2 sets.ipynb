```python
a = set()

a.add(1)
print(a)

a.add(2)
print(a)

a.add(2)
print(a)
```

```python
for element in a:
    print(element)
```

```python
# remove duplicated element
q = [1, 1, 2, 4, 2]
p = set()

for element in q:
    p.add(element) # add element in list to set
print(p)
```

```python
# new list only contain the unique of original list
e = list()
for element in p:
    e.append(element)
print(e)
```

```python
b = set()
b.add("apple")
b.add("banana")
b.add(1)
print(b) # order does not matter in set
```

```python
c = [1, 3, 4, 1, 3]
d = set()
for element in c:
    d.add(element)
print(sum(d))
```
