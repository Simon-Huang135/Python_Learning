```python
# zip: combine things for iterator
# lambda: a simplified function
# map: map the known function and parameter
```

```python
# zip
a = [1, 2, 3]
b = [4, 5, 6]

# 1
zip(a, b) # if just print zip(), that is an object
list(zip(a, b)) # now can show index 0, 1, 2... of a and b both iteratively
for i, j in zip(a, b):
    print(i, j)
    #print(i / 2, j * 2)

# 2
c = list(zip(a, a, b))
for i, j ,k in c:
    print(i, j, k)
```

```python
# function
def func(x, y):
    return x + y

print(func(2, 3))

# lambda
function = lambda x, y: x + y
print(function(2, 3))
```

```python
# map
result = list(map(func, [1, 2, 3], [2, 5, 8])) # if just print map(), that is an object
print(result)            # x        # y
```
