```python
# no function
a = 1
b = 2
c = a + b
print(c)
```

```python
# function
def sumc (x, y):
    return x + y

c = sumc(1, 2)
print(c)
```

```python
# default parameter of function
def sale_car(price, color, brand, is_second_hand):
    print('price:', price, 
          'color:', color,  
          'brand:', brand, 
          'is_second_hand:', is_second_hand,)
    
sale_car(1000, 'red', 'carmy', True)
```

```python
# global & local 
Apple = 100
a = None
def fun():
    a = 20
    return a+100
   
print(Apple)
print('past', a)
print(fun())
print('now', a)

print('')

Apple = 100
a = None
def func():
    global a
    a = 20
    return a+100
   
print(Apple)
print('past', a)
print(func())
print('now', a)
```
