```python
def are_u_sad(is_rainy, has_umbrella):
    return is_rainy and not has_umbrella
```

```python
are_u_sad(True, False) # rainy and you do not has umbrella
```

```python
are_u_sad(True, True) # rainy and you has umbrella
```
