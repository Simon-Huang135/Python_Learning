```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
# read url, if has Chinese, apply decode()
html = urlopen("https://morvanzhou.github.io/static/scraping/list.html").read().decode('utf-8')
print(html)

''' this is the class in css (.someting)
	<style>
	.jan {
		background-color: yellow; \
	}
	.feb {
		font-size: 25px;
	}
	.month {
		color: red;
	}
	</style>
</head>'''
```

```python
# in css, a class can define attributes of the content
soup = BeautifulSoup(html, features = 'lxml')

# choose tag info <h1>
print(soup.h1) # get the tag has <h1>....<h1>
print('')

# choose tag info <p>
print('\n', soup.p) # get the tag has <p>....<p>
print('')

# choose tag info <a href
all_href = soup.find_all('a')
all_href = [l['href'] for l in all_href]
print(all_href)
```

```python
# use class to narrow search
month = soup.find_all('li', {"class": "month"}) # find the li list then find the class month in the li list
print(month) # if only month, the list tag will be shown, need the text

for m in month: 
    print(m.get_text()) # get the text but not tag
    
```

```python
# use class to narrow search
jan = soup.find('ul', {"class": 'jan'}) # find the ul list first, then find the class jan in the ul list
print(jan) # so jan is the entire 
d_jan = jan.find_all('li')              # use jan as a parent
for d in d_jan:
    print(d.get_text())
```

```python
a = soup.find_all('ul', {'class': 'jan'})
print(a)
print('')
for d in a:
    print(d.get_text())
```
