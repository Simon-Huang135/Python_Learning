```python
from urllib.request import urlopen
import re

# read html from web, then use regular expression to match the key info
```

```python
# read html from web
html = urlopen('https://morvanzhou.github.io/static/scraping/basic-structure.html').read().decode('utf-8')
# read web link and if has Chinese, apply decode()
print(html) # get original html code of the weblink
```

```python
# use regular expression to match the key info
res = re.findall(r"<title>(.+?)</title>", html) # find the title of the html
print(res[0])
```

```python
# use regular expression to match the key info
res = re.findall(r"<p>(.*?)</p>", html, flags = re.DOTALL)    # find paragraph of html # re.DOTALL if multi line
print(res[0]) # sometimes the re matching will get error like the tag of link: <a href=... 
                                          # BeatifulSoup can get no error like this
```

```python
# use regular expression to match the key info
res = re.findall(r'href="(.*?)"', html) # find all other link in this html cide
print(res)
```
