```python
import pandas as pd
import numpy as np
```

```python
# Circumscribed circle of rectangle - using two functions
# The diameter of the circle is the diagonal of the rectangle
# so r = d/2, Area = pi * r^2 
# so r = sqrt((power(Length, 2) + power(Height, 2)))/2
# Area = pi * power(sqrt((power(Length) + power(Height)))/2, 2)

def get_circumcircle_area(l, h):
    r = np.sqrt(l ** 2 + h ** 2)/2
    return np.pi * r ** 2

# since we get the data row by row with length and height
def wrapper(row):
    return get_circumcircle_area(row['Length'], row['Height'])


rect = pd.read_excel('Book19.xlsx', index_col = 'ID')
rect['CA'] = rect.apply(wrapper, axis = 1) # when axis = 1, scan row by row

print(rect)
```

```python
# Circumscribed circle of rectangle - using 1 function and lambda
def get_circumcircle_area(l, h):
    r = np.sqrt(l ** 2 + h ** 2)/2
    return np.pi * r ** 2

rects = pd.read_excel('Book19.xlsx', index_col = 'ID')
rects['CA'] = rects.apply(lambda row: get_circumcircle_area(row['Length'], row['Height']), axis = 1) 
# when axis = 1, scan row by row

print(rects)
```
