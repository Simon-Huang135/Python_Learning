```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Standard 
# using plot.bar of matplotlib pyplot
books = pd.read_excel('Book4.xlsx')
# sort by number
books.sort_values(by = 'Number', inplace = True, ascending = False) # inplace would sort in current dataframe
print(books)
# ploting
books.plot.bar(x = 'Field', y = 'Number', color = 'black', title = 'Internetional Students by Field')
# let figure be tight
plt.tight_layout()
plt.show()
```

```python
# Standard
# using matplotlib
books1 = pd.read_excel('Book4.xlsx')
books1.sort_values(by = 'Number', inplace = True, ascending = False) # inplace would sort in current dataframe

# plt setting
plt.bar(books1.Field, books1.Number, color = 'black')
plt.xticks(books1.Field, rotation = 70, ha = 'right')
plt.xlabel('Field')
plt.ylabel('Number')
plt.title('Internetional Students by Field', fontsize = 16)
plt.tight_layout()
plt.show()
```

```python
# Grouping
# using plot.bar of matplotlib pyplot, optimizing using plt (deep optimization)
books2 = pd.read_excel('Book5.xlsx')
books2.sort_values(by = [2016, 2017], inplace = True, ascending = False)
print(books2)

books2.plot.bar(x = 'Field', y = [2016, 2017], color = ['black', 'red']) 
'''
books2.plot.bar(x = 'Field', y = [2016, 2017], color = ['black', 'red'], \
                title = '2016 - 2017 Internetional Students of Field')
'''
# setting
plt.title('2016 - 2017 Internetional Students of Field', fontsize = 16, fontweight = 'bold')
plt.xlabel('Field',fontweight = 'bold')
plt.ylabel('Number',fontweight = 'bold')
ax = plt.gca()
ax.set_xticklabels(books2['Field'], rotation = 70, ha = 'right')
f = plt.gcf()
f.subplots_adjust(left = 0.2, bottom = 0.42)

#plt.tight_layout()
plt.show()
```

```python
# Stacked
# using plot.bar of matplotlib pyplot, optimizing using plot.bar
books3 = pd.read_excel('Book5.xlsx')
# sorting the stacked data
books3['Total'] = books3[2016] + books3[2017]
books3.sort_values(by = 'Total', inplace = True, ascending = True)
print(books3)

#plot horizontally - barh
books3.plot.barh(x = 'Field', y = [2016, 2017], stacked = True)

# setting
plt.title('2016 - 2017 Internetional Students of Field', fontsize = 16, fontweight = 'bold')
plt.xlabel('Field',fontweight = 'bold')
plt.ylabel('Number',fontweight = 'bold')
ax = plt.gca()
ax.set_xticklabels(books3['Field'], rotation = 70, ha = 'right')
f = plt.gcf()
f.subplots_adjust(left = 0.2, bottom = 0.42)

plt.show()
```
