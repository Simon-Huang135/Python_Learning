```python
import pandas as pd
```

```python
s1 = pd.Series([1, 2, 3], index = [1, 2, 3], name = 'A')
s2 = pd.Series([10, 20, 30], index = [1, 2, 3], name = 'B')
s3 = pd.Series([100, 200, 300], index = [1, 2, 3], name = 'C')
```

```python
# creat a table with dictionary
df0 = pd.DataFrame({s1.name:s1, s2.name:s2, s3.name:s3}) # name as row, series as col
print(df0)
```

```python
# creat a table with list
df1 = pd.DataFrame([s1, s2,s3]) # series as row, name as col
print(df1)
```
