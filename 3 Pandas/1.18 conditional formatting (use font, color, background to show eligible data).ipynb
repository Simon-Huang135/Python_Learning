```python
import pandas as pd
import seaborn as sns
```

```python
# conditional fomatting 1:  font color and background color
# red fontcolor - (< 60), green background green font- (top 1 of each test)
def low_score_red(s):
    color = 'red' if s < 60 else 'black'
    return f'color: {color}'

def highest_score_green(col):
    return ['background-color: lime'if s == col.max() else 'blackgrond-color: white' for s in col]

books = pd.read_excel('Book16.xlsx', sheet_name = 'Sheet1', index_col = 'ID')

# .applymap: elementwise
# .apply: column-/row-/table-wise
books.style.applymap(low_score_red, subset = ['Test_1', 'Test_2', 'Test_3']) \
        .apply(highest_score_green, subset = ['Test_1', 'Test_2', 'Test_3'])
```

```python
# conditional fomatting 2: background color level (1)
    # need to use seaborn
    # (color, n_colors=6, reverse=False, as_cmap=False, input='rgb')
color_map = sns.light_palette('seagreen', as_cmap = True)

books1 = pd.read_excel('Book16.xlsx', sheet_name = 'Sheet1', index_col = 'ID')

books1.style.background_gradient(color_map, subset = ['Test_1', 'Test_2', 'Test_3'])
```

```python
# conditional fomatting 2: progress bar (2)
books2 = pd.read_excel('Book16.xlsx', sheet_name = 'Sheet1')
books2.style.bar(color = 'pink', subset = ['Test_1', 'Test_2', 'Test_3'])
```
