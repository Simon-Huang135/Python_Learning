```python
import pandas as pd
import numpy as np
```

```python
# Pivot table - method 1
books = pd.read_excel('Book14.xlsx')
books['Year'] = pd.DatetimeIndex(books.Date).year # or pd.DatetimeIndex(books['Date']).year

# .pivot_table(data, values=None, index=None, columns=None, aggfunc='mean', 
# fill_value=None, margins=False, dropna=True, margins_name='All')[source]
pt1 = books.pivot_table(index = 'Category', columns = 'Year', values = 'Total',aggfunc = np.sum)

print(pt1)
```

```python
# Pivot table - method 1 (group by - manually)
books1 = pd.read_excel('Book14.xlsx')
books1['Year'] = pd.DatetimeIndex(books['Date']).year

# .groupby(by=None, axis=0, level=None, as_index=True, sort=True, 
# group_keys=True, squeeze=False, observed=False, **kwargs)
groups = books1.groupby(['Category', 'Year'])

# get money and numbers
# total money
s = groups['Total'].sum()
# total number of sale
c = groups['ID'].count()

pt2 = pd.DataFrame({'Sum': s, 'Count': c})
print(pt2)
```
