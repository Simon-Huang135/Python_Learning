```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

# normally we read html from web, then use regular expression to match the key info, but sometime get error
# BeautifulSoup can choose the tag infor instead of using re
# so Beautiful can do in this way:
# 1 find url wanted to scrap
# 2 use urlopen to open the website
# 3 read it
# 4 put the read info in BeautifulSoup
# 5 use BeautifulSoup to choose tag info instead of using re
```

```python
# read url, if has Chinese, apply decode()
html = urlopen("https://morvanzhou.github.io/static/scraping/basic-structure.html").read().decode('utf-8')
print(html)
```

```python
# put the read info (html) in BeautifulSoup
soup = BeautifulSoup(html, features = 'lxml') # use BeautifulSoup to eat the html, then decode in lxml way

# choose tag info <h1>
print(soup.h1) # get the tag has <h1>....<h1>

# choose tag info <p>
print('\n', soup.p) # get the tag has <p>....<p>
```

```python
# choose tag info <a
all_href = soup.find_all('a')
print(all_href)

print('')

# to remove the tag <a href, use a for loop to serach href , print things in "...."
all_href = [l['href'] for l in all_href] # this is a for loop： # for l in all_href:
print(all_href)                                                 #    print(l['href'])
```
