```python
import numpy as np
```

```python
a = np.array([10, 20, 30, 40])
b = np.arange(4)

# subtraction, etc. (this is element operation)
c = a - b
print(c)
print('')

# trigonometric functions
c = 10 * np.sin(a)
print(c)
print('')

# comparator
print(b)
print(b < 3)
```

```python
# 2D matrix operation
a = np.array([[1, 1],
              [0, 1]])
b = np.arange(4).reshape(2,2)

print(a)
print('')

print(b)
print('')

#  element operation
c = a * b
print(c)
print('')

# matrix multiplication, not dot product or cross product
c = np.dot(a, b) # or a.dot(b)
print(c) 

# [[1 1]     [[0 1]    
#  [0 1]]     [2 3]]
 
# = [[1*0+1*2  1*1+1*3] = [[2 4]
#    [0*0+1*2  0*1+1*3]    [2 3]]
```

```python
# create a random array
a = np.random.random((2, 4)) # each element is from 0 to 1
print(a)
print('')

# min, max, sum of the matrix
print(np.sum(a))
print('')

print(np.sum(a, axis = 1)) # axis = 1 find the sum row by row, 0 find the sum col by col
print('')

print(np.min(a, axis = 1)) # 1 min of each row, 0 min of each col
print('')

print(np.max(a, axis = 0)) # 1 min of each row, 0 min of each col
```

```python
a = np.arange(14, 2, -1).reshape(3, 4)
print(a)
print('')

# find the index of the min, max element in matrix
print(np.argmin(a))
print(np.argmax(a))
print('')

# mean value of matrix
print(np.mean(a))
print(a.mean()) # or this way, built-in method mean of numpy.ndarray
print(np.average(a)) # or this way
print('')

# median
print(np.median(a)) # axis = 1 median of each row, 0 median of each col
print('')

# accumulation sum
print(np.cumsum(a)) # default [2 2+3 2+3+4....]
print('')           # if axis = 0, row by row horizontally, 1 col by col vertically
    
# difference between each element
print(np.diff(a))
print('')

# find nonzero
print(np.nonzero(a)) # return the indices of the elements that are non-zero.
# so array 1 is row indices, array 2 is col indices
```

```python
# sort the matrix
a = np.sort(a)
print(a)
print('')

# transpose
print(np.transpose(a))
print('')

# transpose mutliplicates nontranspose, not element multiplication
c = (a.T).dot(a)
print(c)
```

```python
a = np.arange(2, 14).reshape(3, 4)
print(a)
a = np.clip(a, 5, 10) # return an array whose values are limited to [min, max]
print(a)              # the vlaue < 5 will become 5, > 10 will become 10
```
