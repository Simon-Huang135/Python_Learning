```python
import numpy as np
```

```python
a = np.arange(12).reshape(3, 4)
print(a)
print('')

# split 3 x 4 matrix to 2 pieces equally, col by col ï¼ˆcontrolling by axis)
b = np.split(a, 2, axis = 1) # or b = np.hsplit(a, 2)
c = np.split(a, 3, axis = 0) # or c = np.vsplit(a, 3)
print(b)
print(c)
print('')
# print(b[0])
```

```python
# not equally split
b = np.array_split(a, 2, axis = 0)
print(b)
print('')

b = np.array_split(a, 3, axis = 1)
print(b)
```
