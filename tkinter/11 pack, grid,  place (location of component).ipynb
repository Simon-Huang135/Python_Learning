```python
import tkinter as tk
```

```python
# pack
window = tk.Tk()
window.geometry('200x200')

tk.Label(window, text = '1').pack(side = 'top')
tk.Label(window, text = '1').pack(side = 'bottom')
tk.Label(window, text = '1').pack(side = 'left')
tk.Label(window, text = '1').pack(side = 'right')

window.mainloop()
```

```python
# grid
window = tk.Tk()
window.geometry('200x200')

for i in range(4):
    for j in range(3):
        tk.Label(window, text = 1).grid(row = i, column = j, padx = 2, pady = 2) # padx/y, ipadx,y
        
window.mainloop()
```

```python
# place
window = tk.Tk()
window.geometry('500x200')

# tk.Label(window, text = 1).place(x = 20, y = 10, anchor = 'nw') # x, y => pixel point, 
#                                                                 # anchor is the pixel point of the picture
canvas = tk.Canvas(window, height=150, width=500)
canvas.grid(row=1, column=1)
image_file = tk.PhotoImage(file='welcome.gif')
image = canvas.create_image(80, 50, anchor='nw', image=image_file)

window.mainloop()
```
