```python
import tkinter as tk
import tkinter.messagebox
```

```python
window = tk.Tk()
window.title('my window')
window.geometry('200x200')

def hit_me():
    # box show info
    tk.messagebox.showinfo(title = 'Hi', message = 'info box')   # return 'ok'
    # box show warning
    tk.messagebox.showwarning(title = 'Hi', message = 'warning')   # return 'ok'
    # box show error
    tk.messagebox.showerror(title = 'Hi', message = 'error')   # return 'ok'
    # box ask question
    tk.messagebox.askquestion(title = 'Hi', message = 'ask question')   # return 'yes' , 'no'
    # box ask yes or no
    tk.messagebox.askyesno(title = 'Hi', message = 'yes or no')   # return True, False
    # box ask ok or cancel
    tk.messagebox.askokcancel(title = 'Hi', message = 'ok or cancel')   # return True, False
    # box ask yes or no or cancel
    tk.messagebox.askyesnocancel(title = "Hi", message = "yes or no or cancel")     # return, True, False, None

# button
tk.Button(window, text = 'hit me', command = hit_me).pack()
window.mainloop()
```
