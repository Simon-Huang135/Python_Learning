```python
import tkinter as tk
```

```python
window = tk.Tk()
window.title('my window')
window.geometry('300x200')

# label
l = tk.Label(window, 
             bg = 'yellow', 
             width = 20, 
             text = 'empty')
l.pack()

# scale 
def print_selection(v):
    l.config(text='you have selected ' + v)

s = tk.Scale(window, 
             label = 'try me', 
             from_= 0, to = 100, 
             orient = tk.HORIZONTAL,
             length = 300, 
             showvalue = 1, 
             tickinterval = 20, 
             resolution = 0.1, 
             command = print_selection)
s.pack()

window.mainloop()
```
