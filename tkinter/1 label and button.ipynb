```python
import tkinter as tk
```

```python
# build the window
window = tk.Tk()
# window title
window.title('My window')
# window size
window.geometry('200x100')

var = tk.StringVar() # the value storing the variable of string when hit

# define label on window
l = tk.Label(window, 
             textvariable = var, 
             bg = 'white', 
             font = ('Arial', 15), 
             width = 15, 
             height = 2
             )
# label position
l.pack() # or l.place in precise location


# define hit_me function
on_hit = False
def hit_me():
    global on_hit
    if on_hit == False:
        on_hit = True
        var.set('Hi, Telina')
    else:
        on_hit = False
        var.set('Hi, Simon')

# define button
b = tk.Button(window,
              text = 'Click me', 
              width = 15, 
              height = 2, 
              command = hit_me)
b.pack()

window.mainloop()
```
