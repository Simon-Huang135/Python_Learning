```python
import pandas as pd
```

```python
page1 = pd.read_excel('Book17.xlsx', sheet_name = 'Page_001')
page2 = pd.read_excel('Book17.xlsx', sheet_name = 'Page_002')
# page 2 follows page 1 (union - pands: append)
    # if reset_index has default preset, the old index will be still there
    # use reset_index (level=None, drop=False, inplace=False, col_level=0, col_fill='')
students = page1.append(page2).reset_index(drop = True)
# or students = pd.concat([page1, page2], axis = 0).reset_index(drop = True)

print(students)
```

```python
# add a new row at the end of students, in dataframe of panda, row/col is a series
    # page1 = pd.read_excel('Book17.xlsx', sheet_name = 'Page_001')
    # page2 = pd.read_excel('Book17.xlsx', sheet_name = 'Page_002')
    # students = page1.append(page2).reset_index(drop = True)
stu = pd.Series({'ID': 41, 'Name': 'Simon', 'Score': 99})
# .append (other, ignore_index=False, verify_integrity=False, sort=None)
students = students.append(stu, ignore_index = True)

print(students.tail(10))
```

```python
# edit existed value in the dataframe students
# change value of index 39, id 40
    # page1 = pd.read_excel('Book17.xlsx', sheet_name = 'Page_001')
    # page2 = pd.read_excel('Book17.xlsx', sheet_name = 'Page_002')
    # students = page1.append(page2).reset_index(drop = True)
students.at[39, 'Name'] = 'Telina'
students.at[39, 'Score'] = 99

# change the whole row (use id here to change create row for covering)
stu1 = pd.Series({'ID': 33, 'Name': 'Tim', 'Score': 99})
# use index here
students.iloc[32] = stu1

print(students.tail(10))
```

```python
# Insert a row into dataframe (between id = 20 and 21, index = 19 ane 20)
# need to use index to insert, slice dataframe to two dataframe, then add the new row, then merge , like sandwich
page1 = pd.read_excel('Book17.xlsx', sheet_name = 'Page_001')
page2 = pd.read_excel('Book17.xlsx', sheet_name = 'Page_002')
students = page1.append(page2).reset_index(drop = True)
stu2 = pd.Series({'ID': 101, 'Name': 'Steph', 'Score': 101})

slice1 = students[:20] # part 1 is from index 0 to 19
print(slice1)

slice2 = students[20:] # part 2 is from index 20 to end
print(slice2)
students = slice1.append(stu2, ignore_index = True).append(slice2).reset_index(drop = True)

print(students)
```

```python
# delete row index 0 - 9
page1 = pd.read_excel('Book17.xlsx', sheet_name = 'Page_001')
page2 = pd.read_excel('Book17.xlsx', sheet_name = 'Page_002')
students = page1.append(page2).reset_index(drop = True)

#students.drop(index = [0, 1, 2], inplace = True)
#students.drop(index = range(0,10), inplace = True)
students.drop(index = students[1:18].index, inplace = True)

print(students)
```

```python
# conditional delete
page1 = pd.read_excel('Book17.xlsx', sheet_name = 'Page_001')
page2 = pd.read_excel('Book17.xlsx', sheet_name = 'Page_002')
students = page1.append(page2).reset_index(drop = True)

for i in range(2,8):
    students['Name'].at[i] = ''
print(students.head(9))
print('')

# delete operation
missing = students.loc[students['Name'] == '']
print(missing)
print('')

stduents = students.drop(index = missing.index, inplace = True)
    # missing. index - DataFrame.index, the index (row labels) of the DataFrame.
print(students)
print('')

# reset_index
students = students.reset_index(drop = True) 
print(students)
```
