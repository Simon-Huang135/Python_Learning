```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Pie chart
books = pd.read_excel('Book6.xlsx', index_col = 'From')

# just need a series, set orientation
# books['2017'].sort_values(ascending = True).plot.pie(fontsize = 8, startangle = -270)
books['2017'].plot.pie(fontsize = 8, counterclock = False, startangle = -270)
# setting
plt.title('Source of Internetional Students', fontsize = 16, fontweight = 'bold')
plt.ylabel('2017', fontsize = 12, fontweight = 'bold')

plt.show()


print(books)
```
