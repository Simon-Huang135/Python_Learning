```python
import pandas as pd
```

```python
# merge table
students = pd.read_excel('Book9.xlsx', sheet_name = 'Students')
print (students)

score = pd.read_excel('Book9.xlsx', sheet_name = 'Score')
print(score)
# table = students.merge(score, on = 'ID') # inner join, bandon the element without ID
table = students.merge(score, how = 'left', on = 'ID').fillna(0) 
    # join the ID number, reserve th element in left sheet and fill NaN to 0
    # if left is not ID, right is ID then chang on to: left on = .., right on = 'ID'
table.Score = table.Score.astype(int) # change the float to int
print(table)
```

```python
# merge table if ID is preset to index
students1 = pd.read_excel('Book9.xlsx', sheet_name = 'Students', index_col= 'ID')
score1 = pd.read_excel('Book9.xlsx', sheet_name = 'Score', index_col= 'ID')

# table = students.merge(score, how = 'left', on = 'ID').fillna(0) 
# or 
table1 = students1.merge(score1, how = 'left', left_on = students1.index, right_on = score1.index).fillna(0) 
table1.Score = table1.Score.astype(int)
print(table1)
```
