```python
import pandas as pd
import numpy as np
```

```python
page1 = pd.read_excel('Book18.xlsx', sheet_name = 'Page_001')
page2 = pd.read_excel('Book18.xlsx', sheet_name = 'Page_002')
# page 2 side by side page 1

students = pd.concat([page1, page2], axis = 1)
# s1 = pd.Series(['a', 'b'])
# s2 = pd.Series(['c', 'd'])
# pd.concat([s1, s2], ignore_index = True)
print(students)
```

```python
# add a new col
students1 = pd.concat([page1, page2]).reset_index(drop = True)
#students1['Age'] = np.repeat(25, len(students1)) # all 25 
#students1['Age'] = 25 # all 25 
students1['Age'] = np.arange(0, len(students1)) 
print(students1.head(5))
```

```python
# delete col
# drop (labels=None, axis=0, index=None, columns=None, level=None, inplace=False, errors='raise'ï¼‰
students1.drop(columns = ['Age', 'Score'], inplace = True)
print(students1.head(5))
```

```python
# insert col
# insert (loc, column, value, allow_duplicates=False)
students1.insert(1, column = 'Foo', value = np.repeat('foo',len(students1)))
print(students1.head(5))
```

```python
# change the col name
# rename (mapper=None, index=None, columns=None, axis=None, copy=True, inplace=False, level=Non)
students1.rename(columns = {'Foo': 'FOO', 'Name': 'NAME'}, inplace = True)
print(students1.head(5))
```

```python
# delete nonsense cell
students1 = pd.concat([page1, page2]).reset_index(drop = True)
students1['ID'] = students1['ID'].astype(float)
students1['Score'] = students1['Score'].astype(float)

# creat empty ID
for i in range(5, 10):
    students1['ID'].at[i] = np.nan
    
for i in range(2, 4):
    students1['Score'].at[i] = np.nan
    
print(students1.head(11))
print('')

students1.dropna(inplace = True) # dropna check all col row by row
print(students1.head(11))
```
