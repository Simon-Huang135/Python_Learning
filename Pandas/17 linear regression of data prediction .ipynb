```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
books = pd.read_excel('Book15.xlsx', dtype = {'Month': str}) # since month was float, so set to str first to process

# linear regression predefine
slope, intercept, r_value, p_value, std_err = linregress(books.index, books.Revenue)
    # r_value: Correlation coefficient                           x             y
    # p_value: Two-sided p-value for a hypothesis test whose null hypothesis is that 
        # the slope is zero, using Wald Test with t-distribution of the test statistic.
        
# linear regression model: y = kx + b
exp = slope * books.index + intercept 

# show slope, r^2, b
print("slope: %f    intercept: %f    R-squared: %f" % (slope, intercept, r_value**2 ))
#print("slope: %f    intercept: %f" % (slope, intercept))
#print("R-squared: %f" % r_value ** 2)

# plot scatter and trend line
plt.scatter(books.index, books.Revenue) # since the x is month string, so use index 0-23 here
plt.plot(books.index, exp, color = 'red', label='fitted line')

# setting 
plt.title('Sales\n' f'y ={slope} * x + {intercept}')
plt.xticks(books.index, books.Month, rotation = 70, ha = 'right') # then label the month on index
plt.tight_layout()
plt.legend()
plt.show()

# predict future value
print(slope * 35 + intercept) # 35 is 2019.12

print('')
print(books)
```
