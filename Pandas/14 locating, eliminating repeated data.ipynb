```python
import pandas as pd
```

```python
books = pd.read_excel('Book12.xlsx')
books.drop_duplicates(subset = 'Name', keep = 'first', inplace = True)
# if multi-col, subset = [ID, Name...]
print(books)
```

```python
# find the location of duplicated row
books1 = pd.read_excel('Book12.xlsx')
dupe = books1.duplicated(subset = 'Name', keep='first') # mark the duplicated row as True
print(dupe)
print(type(dupe)) 
print('')

# check the data has any duplicated data or not
print(dupe.any()) # return true if there is true
print(dupe.all()) # return true if all true
print('')          

# only show the duplicated row
dupe = dupe[dupe == True] # or dupe = dupe[dupe]
print('')

# dupe is a series here
print(dupe)
print(type(dupe)) 
print('')

# the index (row labels) of the series.
print(dupe.index) 
print('')

print(books1.iloc[dupe.index]) # iloc: Purely integer-location based indexing for selection by position.
```
