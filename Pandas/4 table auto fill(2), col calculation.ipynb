```python
import pandas as pd
```

```python
# col by col calculation 1
books = pd.read_excel('Book1.xlsx', index_col = 'ID')
books['Price'] = books['ListPrice'] * books['Discount']
print(books)
```

```python
# col by col calculation 2
books1 = pd.read_excel('Book1.xlsx', index_col = 'ID')
for i in range(5,16): # books1.index:
    books1['Price'].at[i] = books1['ListPrice'].at[i] * books1['Discount'].at[i]
print(books1)
```

```python
# book price increa by 2 (1)
books2 = pd.read_excel('Book1.xlsx', index_col = 'ID')
books2['ListPrice'] = books2['ListPrice'] + 2
print(books2)
```

```python
# book price increa by 2 (2) using lambda or using function

# def add_2(x):
#     return x + 2

books3 = pd.read_excel('Book1.xlsx', index_col = 'ID')
# books3['ListPrice'] = books3['ListPrice'].apply(add_2)
books3['ListPrice'] = books3['ListPrice'].apply(lambda x: x + 2)
print(books3)
```
