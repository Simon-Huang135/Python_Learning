```python
import pandas as pd
```

```python
# data validation : assert
def score_validation(row):
    try:
        assert 0 <= row.Score <=100
    except:
        print(f'# {row.Name} has an invalid score {row.Score}.')

books = pd.read_excel('Book10.xlsx')
books.apply(score_validation, axis = 1) # axis = 0 means from top to down, 1 means from left to right 
                                            # since we check data row by row, so from left to right
print(books)
```

```python
# data validation : if else
def score_validation(row):
    if not 0 <= row.Score <= 100:
        print(f'# {row.Name} has an invalid score {row.Score}.')

books = pd.read_excel('Book10.xlsx')
books.apply(score_validation, axis = 1) # axis = 0 means from top to down, 1 means from left to right 
                                            # since we check data row by row, so from left to right
print(books)
```
