```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# trend line
books = pd.read_excel('Book7.xlsx', index_col = 'Week')
print(books.columns)

# plot trend line, optimizing using plt (deep optimization)
books.plot(y = ['Accssories', 'Bikes', 'Clothing', 'Components'])
plt.title('Sales Weekly Trend', fontsize = 16, fontweight = 'bold')
plt.ylabel('Total', fontsize = 12, fontweight = 'bold')
plt.xticks(books.index, fontsize = 8) # since week become index col
plt.show()
```

```python
# stack area trend line 
books1 = pd.read_excel('Book7.xlsx', index_col = 'Week')
print(books1.columns)

books1.plot.area(y = ['Accssories', 'Bikes', 'Clothing', 'Components'])
plt.title('Sales Weekly Trend', fontsize = 16, fontweight = 'bold')
plt.ylabel('Total', fontsize = 12, fontweight = 'bold')
plt.xticks(books1.index, fontsize = 8) # since week become index col
plt.show()
```

```python
# stack area bar chart 
books1 = pd.read_excel('Book7.xlsx', index_col = 'Week')
print(books1.columns)

books1.plot.bar(y = ['Accssories', 'Bikes', 'Clothing', 'Components'], stacked = True)
plt.title('Sales Weekly Trend', fontsize = 16, fontweight = 'bold')
plt.ylabel('Total', fontsize = 12, fontweight = 'bold')
plt.xticks(books1.index, fontsize = 8) # since week become index col
plt.show()
```
