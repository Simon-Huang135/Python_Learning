```python
import pandas as pd
from datetime import date, timedelta
```

```python
# function for month increament
def add_month(d, md):
    yd = md // 12
    m = d.month + md%12
    if m != 12:
        yd += m //12
        m = m%12
    return date(d.year + yd, m, d.day)
```

```python
# import data with selected area
books = pd.read_excel('Book0.xlsx',skiprows = 3, usecols = 'C:F', index_col = None, 
                      dtype ={'ID': str, 'InStore': str, 'Date': str}) # NaN is float, so set dtype to str first
print(books)  
print(type(books['ID']))
```

```python
# set value to ID col
start = date(2018, 1, 1)
for i in books.index: # or for i in range(len(books))
    books['ID'].at[i] = i + 1 # or books.at[i, 'ID'] = i + 1
    books['InStore'].at[i] = 'Yes' if i%2 == 0 else 'No'
    books['Date'].at[i] = start + timedelta(days = i) # day increament
    #books['Date'].at[i] = date(start.year + i, start.month, start.day) # year increament
    #books['Date'].at[i] = add_month(start, i) # month increament
    
print(books)
```
