```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Scatter plot
books = pd.read_excel('Book8.xlsx')
print(books.columns)
#print(books.shape)
books.plot.scatter(y = 'sqft_living', x = 'price')
plt.show()
```

```python
# histogram
books1 = pd.read_excel('Book8.xlsx')
print(books1.columns)
#print(books.shape)
books1.sqft_living.plot.hist(bins = 100)
plt.xticks(range(0, max(books1.sqft_living), 40), fontsize = 8, rotation = 60, ha = 'right')
plt.title('sqft_living')
plt.show()

books1.price.plot.hist(bins = 100)
plt.xticks(range(min(books1.price), max(books1.price), 50000), fontsize = 8, rotation = 60, ha = 'right')
plt.title('price')
plt.show()
```

```python
# Density chart
books1.sqft_living.plot.kde()
plt.xticks(range(0, max(books1.sqft_living), 40), fontsize = 8, rotation = 60, ha = 'right')
plt.show()
```
