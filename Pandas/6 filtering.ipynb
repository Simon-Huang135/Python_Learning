```python
import pandas as pd
```

```python
# check the score of students who has 18 <= age <= 30, use fuction to filter
def age_18_to_30(a):
    return 18 <= a < 30 # or a >= 18 and a < 30 

def level_a(s):
    return 85 <= s <= 100

books = pd.read_excel('Book3.xlsx',index_col = 'ID')

#books = books.loc[books['Age'].apply(age_18_to_30)].loc[books['Score'].apply(level_a)]
books = books.loc[books.Age.apply(age_18_to_30)].loc[books.Score.apply(level_a)]

# loc would produce a new dataframe, first filter age, use books['Age'] to get the series Age,
# then apply function to filter, the row which satisfies the age_18_to_30 will be placed in to the new dataframe
# 
print(books)
```

```python
# use lambda to filter
books1 = pd.read_excel('Book3.xlsx',index_col = 'ID')

books1 = books.loc[books.Age.apply(lambda x: 18 <= x < 30 )] \
.loc[books.Score.apply(lambda c: 85 <= c <= 100)]


print(books1)
```
