```python
# copy is a module
import copy
```

```python
# this is the operation that make them same
a = [1, 2, 3]
b = a
print(id(a)) # indexing the a location in hard drive
print(id(b)) # the id is same, so they are the same thing, if change b, also change a

b[0] = 5
print(a)

a[2] = 7
print(b)
```

```python
# shallow copy
c = copy.copy(a)
print(id(a))
print(id(c))

c[0] = 9
print(a)
print(c)
```

```python
# special of shallow copy
a = [1, 2, [3, 4]]
d = copy.copy(a)

print(id(a) == id(d)) # a has different id with d
print(id(a[2]) == id(d[2])) # but only the first layer (shallow part) get copy

a[2][0] = 3333 # if a[2] get changed, d[2] get changed
print(d)
```

```python
# if you want the copy that it do not have any connection, deep copy
a = [1, 2, [3, 4]]
d = copy.deepcopy(a)

print(id(a) == id(d))
print(id(a[2]) == id(d[2])) 

a[2][0] = 3333
print(a)
print(d)
```
