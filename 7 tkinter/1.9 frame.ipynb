```python
import tkinter as tk
```

```python
window = tk.Tk()
window.title('my window')
window.geometry('200x200')

# label on window, layer 1
tk.Label(window, text='on the window', bg = 'yellow').pack()

# main frame on window, layer 2
frm = tk.Frame(window)
frm.pack()

# frame left in the main frame, layer 1 of layer 2
frm_l = tk.Frame(frm, )
frm_l.pack(side = 'left')

tk.Label(frm_l, text = 'on the frm_l1').pack()
tk.Label(frm_l, text = 'on the frm_l2').pack()

# frame right in the main frame, layer 2 of layer 2
frm_r = tk.Frame(frm)
frm_r.pack(side='right')

tk.Label(frm_r, text='on the frm_r').pack()

window.mainloop()
```
