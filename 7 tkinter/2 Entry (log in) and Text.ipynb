```python
import tkinter as tk
```

```python
# build the window
window = tk.Tk()
# window title
window.title('My window')
# window size
window.geometry('200x200')

# define entry, e = tk.Entry(window, show="*")
e = tk.Entry(window, show = None) # show = "*"
e.pack()

# function of buttons
def insert_point():
    var = e.get()
    t.insert('insert', var)
    
def insert_end():
    var = e.get()
    #t.insert('end', var)
    t.insert(2.2, var) # insert location (row 2, col 2)
    
# button 1
b1 = tk.Button(window, 
               text = 'insert point', 
               width = 15,
               height = 2, 
               command = insert_point)
b1.pack()

# button 2
b2 = tk.Button(window, 
               text = 'insert end',
               command = insert_end)
b2.pack()

# text
t = tk.Text(window, height = 2)
t.pack()


window.mainloop()
```
