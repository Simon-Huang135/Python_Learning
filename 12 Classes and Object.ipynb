```python
class Robot0:
    def introduce_self(self):
        print("My name is " + self.name)
    def color_self(self):
        print("My color is " + self.color)
    def weight_self(self):
        print("My weight is", self.weight, "kg")
r1 = Robot0()
r1.name = "Tom"
r1.color = "red"
r1.weight = 30

r2 = Robot0()
r2.name = "Jerry"
r2.color = "blue"
r2.weight = 40

r1.introduce_self()
r1.color_self()
r1.weight_self()
r2.introduce_self()
r2.color_self()
r2.weight_self()
```

```python
# this is better, custom constructer
class Robot:
    def __init__(self, name, color, weight): # constructer in python
        self.name = name
        self.color = color
        self.weight = weight   
    def intro_self(self):
        print("My name is " + self.name)

z1 = Robot("Tom", "red", 30)
z2 = Robot("Jerry", "blue", 40)

z1.intro_self()
z2.intro_self()
```

```python
class Person:
    def __init__(self, name, personality, isSitting):
        self.name = name
        self.personality = personality
        self.isSitting = isSitting
    def sit_down(self):
        self.isSitting = True
    def stand_up(self):
        self.isSitting = False        
```
