```python
total = 0
i = 1
while i < 5:
    total += i
    i += 1
print(total)
```

```python
a = [5, 4, 4, 3, 1, -2, -3, 5]
total1 = 0
i = 0
while i < len(a) and a[i] > 0:
    total1 += a[i]
    i += 1
print(total1) # break up to -2

total2 = 0
for j in range(len(a)):
    if a[j] > 0:
        total2 += a[j]
    else:
        break # break when a[j] < 0
print(total2) # sum up all postive int
```

```python
b = [7, 5, 4, 4, 3, 1, -2, -3, -5, -7]
total = 0
for element in b:
    if element < 0:
        total += element
print(total)
```

```python
total1 = 0
j = len(b) - 1
while b[j] < 0:
    total1 += b[j]
    j -= 1
print(total1)
print(len(b))
```
