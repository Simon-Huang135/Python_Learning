```python
# create a calculator class
class Calculator():
    # class attributes can be done by using __init_
    def __init__(self, name, price, height, width, weight): # if the value put here, it would become default value
        self.name = name # link the parameters into the class attributes
        self.price = price
        self.height = height
        self.width = width
        self.weight = weight
    # function
    def add(self, x, y): # self is a default parameter which will pass to the function one by one
        result = x + y
        print(result)
    def minus(self, x, y):
        result = x - y
        print(result)
    def times(self, x, y):
        result = x * y
        print(result)
    def divide(self, x, y):
        result = x / y
        print(result)
```

```python
# now create an object belongs to class calculator, so the object links to the attributes and function
cal = Calculator('bad calculator', 18, 10, 20, 0.5)

print(cal.name) # call atrributes of cal which is thre object cal links to class
print(cal.price)
print(cal.height)
print(cal.width)
print(cal.weight)

cal.add(3, 4) # call the add funtion in cal, the cal is the object links to class
cal.minus(5, 6)
cal.times(30, 69)
cal.divide(40, 3)
```
