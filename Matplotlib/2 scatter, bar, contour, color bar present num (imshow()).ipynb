```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# scatter plot
n = 1024 # point num
x = np.random.normal(0, 1, n) # mean 0, std 1
y = np.random.normal(0, 1, n)
colorp = np.arctan2(y, x) # color value of each point

print(x, y)

plt.scatter(x, y, s =75, c = colorp, alpha = 0.5)
plt.xlim((-1.5, 1.5))
plt.ylim((-1.5, 1.5))
plt.xticks(()) # hide ticks
plt.yticks(())

plt.show()
```

```python
# bar chart
n = 12 # number of data
x0 = np.arange(n) # 0 1 2 3 ... 12
y1 = (1 - x0 / float(n)) * np.random.uniform(0.5, 1.0, n) # uniform distribution
y2 = (1 - x0 / float(n)) * np.random.uniform(0.5, 1.0, n)

print(y1)
print(-y2)

plt.bar(x0, + y1, facecolor = '#9999ff', edgecolor = 'white')
plt.bar(x0, - y2, facecolor = '#ff9999', edgecolor = 'white')

# add annotation
for x, y in zip(x0, y1): # zip: x, y1 import to x, y respectively, ha/va: horizontal/vertical alignment
    # position of text, import y value into text and remain 2 decimal
    plt.text(x + 0.4, y + 0.05, '%.2f' %y, ha = 'center', va = 'bottom')
    
for x, y in zip(x0, y2): # zip: x, y1 import to x, y respectively, ha/va: horizontal/vertical alignment
    # position of text, import y value into text and remain 2 decimal
    plt.text(x + 0.4, - y - 0.05, '%.2f' %y, ha = 'center', va = 'top')
    
plt.xlim(-1, n)
plt.xticks(()) # hide ticks
plt.ylim(-1.25, 1.25)
plt.yticks(())

plt.show()
```

```python
# contour
def f(x,y):
    # the height function
    return (1 - x / 2 + x**5 + y**3) * np.exp(-x**2 -y**2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n) # each x, y will produce a height
X_c, Y_c = np.meshgrid(x, y) # the X_c and Y_c are the input of mshgrid

# use plt.contourf to filling contours color
# X_c, Y_c and value for (X_c, Y_c) point
plt.contourf(X_c, Y_c, f(X_c, Y_c), 8, alpha = 0.2, cmap = plt.cm.cool) # y in hot/cool map present colors
            # x,  y,   z, number of contour, 8 - 10 
    
# use plt.contour to add contour lines
C = plt.contour(X_c, Y_c, f(X_c, Y_c),8, colors = 'black')
            # x,  y,   z
    
# adding label to present the number of 
plt.clabel(C, inline = True, fontsize = 10)

plt.show()
```
