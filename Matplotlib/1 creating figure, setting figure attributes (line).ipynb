```python
# plotting soure
x = np.linspace(-3, 3, 50)
y1 = 2 * x + 1
y2 = x**2

# creating figure 1
plt.figure(num = 1, figsize = (8,5)) # the content below this would be figure 1
plt.plot(x, y2)
plt.plot(x, y1, color = 'red', linewidth = 1.0, linestyle = '--')

# setting figure 1 attributes
# title
plt.title('Figure')
# axis limit
plt.xlim((-1, 2))
plt.ylim((-2, 3))
# axis label
plt.xlabel('x')
plt.ylabel('y')
# axis ticks
x_ticks = np.linspace(-1, 2, 5) # x ticks point can be created as linspace
plt.xticks(x_ticks, rotation = 15, ha = 'right')
plt.yticks([-2, -1.8, -1, 1.22, 3],
           [r'$really\ bad\ \alpha$', r'$bad$', r'$normal$', r'$good$', r'$really\ good$']
           , rotation = 15, ha = 'right')
# move x and y axis
ax = plt.gca() 
ax.spines['right'].set_color('none') # spines means the 4 sides of the figure, set right spines to dissapear
ax.spines['top'].set_color('none') # set top spines to dissapear
ax.xaxis.set_ticks_position('bottom') # set the bottom axis to be the x axis 
ax.yaxis.set_ticks_position('left') # set the left axis to be the y axis 
ax.spines['bottom'].set_position(('data', 0)) # x axis now is set to y = 0 (between normal and good)
ax.spines['left'].set_position(('data', 0)) # y axis now is set to x = 0


plt.show() # has to use in script
```

```python
# plotting soure
x = np.linspace(-3, 3, 50)
y1 = x**2
y2 = 2 * x + 1

# creating figure 1
plt.figure(num = 1, figsize = (8,5)) # the content below this would be figure 1
l1, = plt.plot(x, y1)
l2, = plt.plot(x, y2, color = 'red', linewidth = 1.0, linestyle = '--') # or put label = 'down' here


# setting figure 1 attributes
# title
plt.title('Figure 1')
# legend (l1, must has a , following for handles)
plt.legend(handles = [l1, l2,], labels = ['up', 'down'], loc = 'best')
# or plt.figlegend(handles = (l1, l2,), labels = ('up', 'down'), loc = 'right')
# axis limit
plt.xlim((-1, 2))
plt.ylim((-2, 3))
# axis label
plt.xlabel('x')
plt.ylabel('y')
# axis ticks
x_ticks = np.linspace(-1, 2, 5) # x ticks point can be created as linspace
plt.xticks(x_ticks)
plt.yticks([-2, -1.8, -1, 1.22, 3],
           [r'$really\ bad$', r'$bad$', r'$normal$', r'$good$', r'$really\ good$'])
# axis position
ax = plt.gca() 
ax.spines['right'].set_color('none') # spines means the 4 sides of the figure, set right spines to dissapear
ax.spines['top'].set_color('none') # set top spines to dissapear
ax.xaxis.set_ticks_position('bottom') # set the bottom axis to be the x axis 
ax.spines['bottom'].set_position(('data', 0)) # x axis now is set to y = 0 (between normal and good)
ax.yaxis.set_ticks_position('left') # set the left axis to be the y axis 
ax.spines['left'].set_position(('data', 0)) # y axis now is set to x = 0

# annotation in figure 1
# add a point on the l1 first, then add a vertical line beneath
x0 = 1
y0 = x0**2

plt.scatter(x0, y0, s = 50, color = 'blue') # the point on l1, s -> size
plt.plot([x0, y0], [y0, 0], 'k--', linewidth = 2) # the vertical line beneath
# method 1 of annotation
plt.annotate(r'$x^2 = %s$'%y0, xy = (x0, y0), xycoords = 'data', xytext = (+30, -30), textcoords = 'offset points',
             fontsize = 16, arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3, rad = .3'))
# method 2 of annotation
plt.text(0.3, -1.2, r'$This\ is\ some\ text.\ \mu\ \sigma_i\ \alpha_t$', fontdict = {'size': 16, 'color': 'r'})






plt.show() # has to use in script
```

```python
# plotting soure
x = np.linspace(-3, 3, 50)
y = 0.1 * x

# creating figure 1
plt.figure() # the content below this would be figure 1
plt.plot(x, y, linewidth = 10) # or put label = 'down' here

# setting figure 1 attributes
# axis limit
plt.ylim(-2, 2)
# axis position
ax = plt.gca() 
ax.spines['right'].set_color('none') # spines means the 4 sides of the figure, set right spines to dissapear
ax.spines['top'].set_color('none') # set top spines to dissapear
ax.xaxis.set_ticks_position('bottom') # set the bottom axis to be the x axis 
ax.spines['bottom'].set_position(('data', 0)) # x axis now is set to y = 0 (between normal and good)
ax.yaxis.set_ticks_position('left') # set the left axis to be the y axis 
ax.spines['left'].set_position(('data', 0)) # y axis now is set to x = 0

# set ticks visibility
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(12)
    label.set_bbox(dict(facecolor = 'red', edgecolor = 'None', alpha = 0.5))

plt.show() # has to use in script
```
