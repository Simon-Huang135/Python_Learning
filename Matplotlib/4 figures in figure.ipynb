```python
import matplotlib.pyplot as plt
```

```python
# figures in figure

fig = plt.figure()
x = [1, 2, 3, 4, 5, 6, 7]
y = [1, 3, 4, 2, 5, 8, 6]

# below are all percentage
left, bottom, width, height = 0.1, 0.1, 0.8, 0.8
ax1 = fig.add_axes([left, bottom, width, height])  # main figure
ax1.plot(x, y, 'r')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('title')

# add figure 1
ax2 = fig.add_axes([0.2, 0.55, 0.25, 0.25])  # figure inside 1
ax2.plot(y, x, 'b')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('title inside 1')

# different method to add figure 2
plt.axes([0.63, 0.2, 0.25, 0.25]) # figure inside 2
plt.plot(y[: : 1], x, 'g') # reverse the y value here
plt.xlabel('x')
plt.ylabel('y')
plt.title('title inside 2')

# different method to add figure 3
plt.axes([0.3, 0.2, 0.25, 0.25]) # figure inside 3
plt.plot(y[: : -1], x, 'g') # reverse the y value here
plt.xlabel('x')
plt.ylabel('y')
plt.title('title inside 2')

plt.show()
```
